```python
from chess import (
                Chessboard,
                king_rule_attack, 
                knight_rule_attack,
                wbishop_rule_attack,
                bbishop_rule_attack
                )

from abt_chess import ABTChess

%matplotlib inline
```

```python
chess_experiment = Chessboard(
    6,
    [
        ('knight', 4, (255, 0, 0), 2, knight_rule_attack),
        ('wbishop', 3, (255, 255, 255), 1, wbishop_rule_attack),
        ('bbishop', 2, (0, 0, 150), 3, bbishop_rule_attack),
    ]
)
chess_experiment.show_field()
```

```python
abt = ABTChess(chess_experiment)
chess_experiment.show_field()
abt.run_colorization()
```
